# LIFE EXPECTANCY GDP PORTFOLIO

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
df = pd.read_csv("all_data.csv")
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.Country.unique()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.boxplot(x="Country", y="Life expectancy at birth (years)", data = df)
ax.set_xticks(range(len(df.Country.unique())))
ax.set_xticklabels(df.Country.unique())
plt.savefig("Boxplot.jpg")
plt.show()
plt.clf()
```

# Insight from the boxplot above
Zimbabwe has the least life expentancy at birth than any other country. THe gap is huge, from the data. The average life expectancy is approximately 47 for Zimbabwe. In the other countries, the life expectancy is higher. For instance, Chile's life expentancy at birth is not that diffent from that of Germany but definately higher than China, Mexico. There is no differnce in that of Chile and United States of America.

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(x="Year", y="Life expectancy at birth (years)", hue ="Country", data = df)
plt.savefig("Scatter Plot.jpg")
plt.show()
plt.clf()
```

# Insight from the Scatter Plot of *Year*, *Life Expectancy* colored by Country.
From the scatter plot above, the life expectancy as birth for Zimbabwe decreased from before **2000** and started increasing after **2004**. Zimbabwe appears to be the country that has seen a significant growth in their life expectancy at birth. The other countries appears to be increasing at a constant rate except for Mexico whose life expectancy decreased between **2006 and 2008**  

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(x = "Year", y= "GDP", hue = "Country", data = df)
plt.savefig("Scatter of GDP.jpg")
plt.show()
plt.clf()
```

# Insight from the scatter plot of *GDP* and *Year* colored by Country
The GDP for United State of America appears to be the highest and increases over the years. Somewhere between 2008 and 2010, the GDp growth for USA decreased and kept increasing afterwards. For China, they started at a very low GDP adn increasingly passed all the other counries except for USA. China's GDP growth has taken a steep slope shape. Zimbabwe appears to be the countries with the least GDP and there is no significant increase in over the years. Germany's GDP growth took a sinosodal shape after **2008**. It keeps increasing and decreasing at the same rate.

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(y = "GDP", x= "Life expectancy at birth (years)", hue = "Country", data = df)
plt.savefig("Scatter of GDP_Life expectancy at birth_years.jpg")
plt.show()
plt.clf()
```

# Insight from the Scatter plot of GDP and Life expectancy at birth (years)
From the graph, the life expectancy for a country with a very good growing GDP, appears not to be increasing significantly than for a country that has a stagnant GDP.
